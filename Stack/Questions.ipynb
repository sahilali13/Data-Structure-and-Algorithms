## Use a single python list to implement three stacks


```python
class MultiStack:
    def __init__(self, stack_size):
        self.numberStack = 3
        self.custList =  [None] * (stack_size * self.numberStack)
        self.sizes = [0] * self.numberStack
        self.stack_size = stack_size
    
    def isFull(self, stackNumber):
        if self.sizes[stackNumber] == self.stack_size:
            return True
        else:
            return False
    
    def isEmpty(self, stackNumber):
        if self.sizes[stackNumber] == 0:
            return True
        else:
            return False

    def indexOfTop(self, stackNumber):
        offset = stackNumber * self.stack_size
        return offset + self.sizes[stackNumber] - 1

    def push(self,stackNumber, value):
        if self.isFull(stackNumber = stackNumber):
            return False
        else:
            self.sizes[stackNumber] += 1
            self.custList[self.indexOfTop(stackNumber = stackNumber)] = value
    
    def pop(self, stackNumber):
        if self.isEmpty(stackNumber = stackNumber):
            return "None"
        else:
            value = self.custList[self.indexOfTop(stackNumber = stackNumber)]
            self.custList[self.indexOfTop(stackNumber = stackNumber)] = None
            self.sizes[stackNumber] -= 1
            return value
    
    def peek(self, stackNumber):
        if self.isEmpty(stackNumber = stackNumber):
            return "Empty"
        else:
            value = self.custList[self.indexOfTop(stackNumber = stackNumber)]
            return value
```

## Create Stack with min method


```python
class Node:
    def __init__(self, value = None, next = None, min = None):
        self.value = value
        self.next = next
        self.min = min
    
    def __str__(self):
        return str(self.value)
    
class LinkedList:
    def __init__(self):
        self.head = None
        self.min = None
    
    def __iter__(self):
        curNode = self.head
        while curNode:
            yield curNode
            curNode = curNode.next

class MinStack:
    def __init__(self):
        self.ll = LinkedList()

    def __str__(self):
        values = [str(_) for _ in self.ll]
        return " <- ".join(values)
    
    def min(self):
        if self.ll.min:
            return self.ll.min
        else:
            return None
    
    def isEmpty(self):
        if self.ll.head is None:
            return True
        else:
            return False
        
    def push(self, value):
        if (self.ll.min is None) or (value < self.ll.min):
            min = value
            self.ll.min = min
        else:
            min = self.ll.min
        
        new_node = Node(value = value, min = min)

        if self.ll.head is None:
            self.ll.head = new_node
        else:
            new_node.next = self.ll.head
            self.ll.head = new_node

    def pop(self):
        if self.isEmpty():
            return None
        else:
            popped_node_value = self.ll.head.value
            self.ll.head = self.ll.head.next
            self.ll.min = self.ll.head.min
            return popped_node_value
            
```

## Stack of Plates (Set of Stacks)


```python
class SetOfStack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.stacks = list()

    def __str__(self):
        if self.stacks == []:
            return "None"
        else:
            result = ""
            for _ in self.stacks:
                for __ in _:
                    result += str(__) + " "
                result += ' | '
            return result
    
    def push(self, value):
        if self.stacks == [] or len(self.stacks[-1]) == self.capacity:
            self.stacks.append([value])
        else:
            self.stacks[-1].append(value)
    

    def pop(self):
        if self.stacks == []:
            return None
        else:
            popped_node = self.stacks[-1].pop()
            if self.stacks[-1] == []:
                self.stacks.pop()
            return popped_node
        
    def popAt(self, index):
        if self.stacks == [] or index >= len(self.stacks) or index < 0:
            return None
        else:
            return self.stacks[index].pop()
```

## Queue via Stacks


```python
class Node:
    def __init__(self, value = None):
        self.value = value
        self.next = None

    def __str__(self):
        return str(self.value)
    
class LinkedList:
    def __init__(self):
        self.head = None

    def __iter__(self):
        curNode = self.head
        while curNode:
            yield curNode
            curNode = curNode.next

class CustomStack:
    def __init__(self):
        self.linkedlist = LinkedList()
    
    def __str__(self):
        if self.isEmpty():
            return "None"
        else:
            values = [str(_) for _ in self.linkedlist]
            return " <- ".join(values)

    def isEmpty(self):
        if self.linkedlist.head:
            return False
        return True
    
    def push(self, value):
        new_node = Node(value = value)
        if self.isEmpty():
            self.linkedlist.head = new_node
        else:
            new_node.next = self.linkedlist.head
            self.linkedlist.head = new_node

    def pop(self):
        if self.isEmpty():
            return None
        else:
            popped_node_value = self.linkedlist.head
            self.linkedlist.head = self.linkedlist.head.next
            return popped_node_value

class customQueue:
    def __init__(self):
        self.stack1 = CustomStack()
        self.stack2 = CustomStack()
        
    def __str__(self):
        if self.stack1.isEmpty():
            return "None"
        else:
            values = [str(_) for _ in self.stack1.linkedlist]
            return " <- ".join(values)
    
    def enqueue(self, value):
        self.stack1.push(value = value)
    
    def dequeue(self):
        if self.stack1.isEmpty():
            return None
        else:
            while not self.stack1.isEmpty():
                self.stack2.push(value = self.stack1.pop())
            popped_node_value = self.stack2.pop()
            while not self.stack2.isEmpty():
                self.enqueue(value = self.stack2.pop())
            return popped_node_value
```

## Route between nodes


```python
class Graph:
    def __init__(self, gdict: dict = None) -> None:
        if gdict is None:
            gdict = {}
        self.gdict = gdict
    
    def add_edge(self, vertex, edge) -> None:
        self.gdict[vertex].append(edge)
    
    def check_route(self, startNode, endNode) -> bool:
        visited = [startNode]
        queue = [startNode]
        
        while queue:
            deVertex = queue.pop(0)
            for adjacentVertex in self.gdict[deVertex]:
                if adjacentVertex not in visited:
                    if adjacentVertex == endNode:
                        return True
                    visited.append(adjacentVertex)
                    queue.append(adjacentVertex)
                    
        return False

if __name__ == '__main__':
    customDict = {  'a': ['c', 'd', 'b'],
                    'b': ['j'],
                    'c': ['g'],
                    'd': [],
                    'e': ['f', 'a'],
                    'f': ['i'],
                    'g': ['d', 'h'],
                    'h': [],
                    'i': [],
                    'j': []
                }
    
    g = Graph(gdict = customDict)
    print(g.check_route(startNode = 'a', endNode = 'j'))
```

## Minimal Binary Search Tree


```python
class BSTNode:
    def __init__(self, data=None, left = None, right= None) -> None:
        self.data = data
        self.left = left
        self.right = right
    
    def display(self) -> None:
        lines, *_ = self._display_aux()
        for line in lines:
            print(line)

    def _display_aux(self):
        """Returns list of strings, width, height, and horizontal coordinate of the root."""
        # No child.
        if self.right is None and self.left is None:
            line = '%s' % self.data
            width = len(line)
            height = 1
            middle = width // 2
            return [line], width, height, middle

        # Only left child.
        if self.right is None:
            lines, n, p, x = self.left._display_aux()
            s = '%s' % self.data
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
            second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
            shifted_lines = [line + u * ' ' for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

        # Only right child.
        if self.left is None:
            lines, n, p, x = self.right._display_aux()
            s = '%s' % self.data
            u = len(s)
            first_line = s + x * '_' + (n - x) * ' '
            second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
            shifted_lines = [u * ' ' + line for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

        # Two children.
        left, n, p, x = self.left._display_aux()
        right, m, q, y = self.right._display_aux()
        s = '%s' % self.data
        u = len(s)
        first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
        second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
        if p < q:
            left += [n * ' '] * (q - p)
        elif q < p:
            right += [m * ' '] * (p - q)
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2

def minimalTree(sortedArray):

    if sortedArray:
        lenArray = len(sortedArray)
        
        mid = lenArray // 2
        
        return BSTNode(data = sortedArray[mid], 
                         left = minimalTree(sortedArray = sortedArray[:mid]), 
                         right = minimalTree(sortedArray = sortedArray[mid + 1:]))
    else:
        return None



if __name__ == '__main__':
    sortedArray = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    bst = minimalTree(sortedArray = sortedArray)
    bst.display()


    
```

## List of depths


```python
class LinkedList:
    def __init__(self, val) -> None:
        self.val = val
        self.next = None
    
    def add(self, val) -> None:
        if self.next == None:
            self.next = LinkedList(val)
        else:
            self.next.add(val)

    def __str__(self) -> str:
        return "({val})".format(val = self.val) + str(self.next)

class BinaryTree:
    def __init__(self, val) -> None:
        self.val = val
        self.left = None
        self.right = None
    
def depth(tree: BinaryTree) -> int:
    if tree == None:
        return 0
    if tree.left == None  and tree.right == None:
        return 1
    else:
        depthLeft = 1 + depth(tree.left)
        depthRight = 1 + depth(tree.right)
        if depthLeft > depthRight:
            return depthLeft
        else:
            return depthRight
 
def treeToLinkedList(tree: BinaryTree, custDict = {}, d = None):
    if d == None:
        d = depth(tree)
    if custDict.get(d) == None:
        custDict[d] = LinkedList(tree.val)
    else:
        custDict[d].add(tree.val)
        if d == 1:
            return custDict
    if tree.left != None:
        custDict = treeToLinkedList(tree.left, custDict, d-1)
    if tree.right != None:
        custDict = treeToLinkedList(tree.right, custDict, d-1)
    return custDict

if __name__ == '__main__':
    mainTree = BinaryTree(val = 1)
    two = BinaryTree(val = 2)
    three = BinaryTree(val = 3)
    four = BinaryTree(val = 4)
    five = BinaryTree(val = 5)
    six = BinaryTree(val = 6)
    seven = BinaryTree(val = 7)

    mainTree.left = two
    mainTree.right = three

    two.left = four
    two.right = five

    three.left = six
    three.right = seven

    customDict = treeToLinkedList(tree = mainTree)

    for depthLevel, linkedList in customDict.items():
        print("{0} {1}".format(depthLevel, linkedList))
```

## Check Balanced - LeetCode 110


```python
class Node:
    def __init__(self, value, left = None, right = None) -> None:
        self.value = value
        self.left = left
        self.right = right

def getHeight(root: Node) -> int:
    if root is None:
        return 0
    return 1 + max(getHeight(root = root.left), getHeight(root = root.right))

def isBalanced(root: Node) -> bool:
    if abs(getHeight(root = root.left) - getHeight(root = root.right)) <= 1:
        return True
    return False

if __name__ == '__main__':
    N1 = Node(value = 'N1')
    N2 = Node(value = 'N2')
    N3 = Node(value = 'N3')
    N4 = Node(value = 'N4')
    N5 = Node(value = 'N5')
    N6 = Node(value = 'N6')
    N7 = Node(value = 'N7')
    N8 = Node(value = 'N8')
    

    N1.left = N2
    N1.right = N3

    N2.left = N4
    N2.right = N5

    N3.right = N6
    N6.right = N7
    N7.right = N8
    

    print(isBalanced(root = N1))
```

## Validate BST - LeetCode 98


```python
class TreeNode:
     def __init__(self, value) -> None:
         self.val = value
         self.left = None
         self.right = None

def isValidBST(root: TreeNode) -> bool:
    if root is None:
        return True
    if root.left is None and root.right is None:
        return True
    if root.val < root.left.val or root.val > root.right.val:
        return False
    return isValidBST(root = root.left) and isValidBST(root = root.right)

if __name__ == '__main__':
    
    A1 = TreeNode(value = 4)
    A2 = TreeNode(value = 1)
    A3 = TreeNode(value = 3)
    A4 = TreeNode(value = 2)
    A5 = TreeNode(value = 5)
    

    A1.left = A2
    A1.right = A3

    A3.left = A4
    A3.right = A5

    print(isValidBST(root = A1))
```

## In-order Successor in BST - LeetCode 285


```python
class Node: 
    def __init__(self, key) -> None: 
        self.data = key 
        self.left = None
        self.right = None

def insert(node: Node, data):
    if node is None:
        return Node(data)
    else:
        if data <= node.data:
            temp = insert(node.left, data)
            node.left = temp
            temp.parent = node
        else:
            temp = insert(node.right, data)
            node.right = temp
            temp.parent = node      
        return node

def minimumValue(node: Node) -> Node:
    current = node
    while current is not None:
        if current.left is None:
            break
        current = current.left
    return current

def inOrderSuccessor(root, n: Node) -> Node:
    if n.right is not None:
        return minimumValue(node = n.right)
    else:
        p = n.parent
        while p is not None:
            if n != p.right:
                break
            n = p
            p = p.parent
        return p

if __name__ == '__main__':
    root = None
    root = insert(node = root, data = 4)
    root = insert(node = root, data = 2)
    root = insert(node = root, data = 8)
    root = insert(node = root, data = 1)
    root = insert(node = root, data = 3)
    root = insert(node = root, data = 5)
    root = insert(node = root, data = 9)

    successor = inOrderSuccessor(root = root, n = root.right)
    print(successor.data)
```

## Build Order


```python
def createGraph(projects: list, dependencies: list) -> dict:
    projectGraph = {}
    for project in projects:
        projectGraph[project] = []
    for pairs in dependencies:
        projectGraph[pairs[0]].extend(pairs[1])
    return projectGraph

project = ['a', 'b', 'c', 'd', 'e', 'f']
dependencies = [('a','d'), ('f','b'), ('b','d'), ('f','a'), ('d','c')]

def findDependencies(graph: dict) -> list:
    setD = set()
    for key in graph:
        for value in graph:
            if key in graph[value]:
                setD.add(key)
    return list(setD)


def findBuildOrder(projects: list, dependencies: list) -> list:
    customGraph = createGraph(projects = projects, dependencies = dependencies)
    order = []
    while customGraph:
        nodesWithDependencies = findDependencies(graph = customGraph)
        for project in projects:
            if project not in nodesWithDependencies:
                customGraph.pop(project)
                projects.remove(project)
                order.append(project)
    return order

if __name__ == '__main__':
    customGraph = createGraph(projects = project, dependencies = dependencies)
    print(findBuildOrder(projects = project, dependencies = dependencies))
```

# First Common Ancestor - LeetCode 236


```python
class Node:
    def __init__(self, value, left = None, right = None) -> None:
        self.value = value
        self.right = right
        self.left = left

def search(root: Node, n: Node) -> bool:
    if root is None:
        return False
    if root.value == n.value:
        return True
    return (search(root = root.left, n = n) or search(root = root.right, n = n))
    
def findFirstCommonAncestor(n1: Node, n2: Node, root: Node) -> Node:
    checkFound1 = search(root = root.left, n = n1) and search(root = root.right, n = n2)
    checkFound2 = search(root = root.left, n = n2) and search(root = root.right, n = n1)

    isFound = checkFound1 or checkFound2

    if isFound:
        return root
    else:
        if search(root = root.left, n = n1) and search(root = root.left, n = n2):
            return findFirstCommonAncestor(n1 = n1, n2 = n2, root = root.left)
        else:
            return findFirstCommonAncestor(n1 = n1, n2 = n2, root = root.right)

if __name__ == '__main__':
    
    N55 = Node(value = 55)
    N44 = Node(value = 44)
    N77 = Node(value = 77)
    N22 = Node(value = 22)
    N99 = Node(value = 99)
    N35 = Node(value = 35)
    N88 = Node(value = 88)
    N90 = Node(value = 90)
    N95 = Node(value = 95)
    N54 = Node(value = 54)
    N33 = Node(value = 33)

    N55.left = N44
    N55.right = N77

    N44.left = N22
    N44.right = N99
    
    N22.left = N35
    N22.right = N88

    N99.left = N90
    N99.right = N95

    N88.left = N54

    N90.right = N33

    print(findFirstCommonAncestor(n1 = N33, n2 = N88, root = N55))
```
